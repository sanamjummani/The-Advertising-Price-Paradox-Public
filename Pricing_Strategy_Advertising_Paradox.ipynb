3. Calculate and report the price elasticity for when the brand it is not featured vs. when it is featured. Discuss how advertising affects sales and price elasticity.

When Tropicana is not featured, the price elasticity of demand (PED) is given by the coefficient on log(price): −2.04.
When Tropicana is featured, the PED is the sum of the coefficient on log(price) and the interaction term log(price)×feat: −2.04 + (−1.46) = −3.50.
So, when featured, Tropicana’s demand becomes more elastic - it's elasticity drops by about 1.5, moving from −2.04 to −3.50. Advertising both raises sales levels and increases price sensitivity. We can conclude that when Tropicana is advertised, customers buy more overall, but their quantity demanded responds more strongly to price changes than it would without advertising.

4. Use the optimal pricing formula (on slide 6 of lecture 6) to calculate the optimal retail price (per Carton) for Tropicana when the it is not advertised vs when it is advertised. Discuss how advertising impacts the brand’s pricing strategy.

Profit-maximizing price for OJ given PED of $\gamma$:

price = ($\gamma$ / (1 + $\gamma$)) * cost

From the above calculations, we found that:

$\gamma$ when not advertised = -2.0

$\gamma$ when advertised = -3.5


Solving for optimal retail price when not advertised:
price = (-2.0 / (1 - 2.0)) * cost
      = (-2 / -1) * cost
      = 2cost
Solving for optimal retail price when advertised:
price = -3.5 / (1 - 3.5)) * cost
      = (-3.5 / -2.5) * cost
      = 1.4cost
Example
If cost = $1,
optimal price when not advertised = $2.00
optimal price when advertised = $1.40
When Tropicana is not advertised, demand is less elastic (−2.0), and Tropicana would charge a higher markup of about about 2× cost to maximize profit. So if, for example, cost is 1.00 USD, optimal price would be 2.00 USD
When Tropicana is advertised, demand becomes more elastic (−3.50), and Tropicana would charge a lower markup of about 1.40× cost to maximize profit. So if, for example, cost is 1.00 USD, optimal price would be 1.40 USD
Therefore, advertising increases sales but also increases consumers’ price sensitivity, so the profit-maximizing price is lower in advertised weeks. The brand should advertise and reduce price simultaneously to maximize profit when advertising.

5. Create three new variables variables (price_int, price_decimal1, price_decimal2) to represent the three digits of price in unit of dollars. For example, for price = 2.87, price_int = 2, price_decimal1 = 0.8, and price_decimal2 = 0.07.

```python
df["price_int"] = np.floor(df["price"])
df["price_decimal1"] = (np.floor(df["price"]*10) % 10) / 10
df["price_decimal2"] = (np.floor(df["price"]*100) % 10) / 100

# Look at the some sample rows
print(df[["price", "price_int", "price_decimal1", "price_decimal2"]].sample(10))
```

```python
# Frequency tables
print(df["price_int"].value_counts().sort_index())
print(df["price_decimal1"].value_counts().sort_index())
print(df["price_decimal2"].value_counts().sort_index())
```

price_int: Prices cluster at $1, $2, and $3, which is typical for orange juice
price_decimal1 (first decimal digit): The distribution is not uniform. Values around 0.8–0.9 appear far more often, showing that stores frequently price items at .79, .89, .99, etc., rather than round decimals like .10 or .20.
price_decimal2 (second decimal digit): The distribution is extremely skewed. 0.09 dominates, indicating that most prices end in 9 cents. Other digits occur far less often.

Interpretation: The results suggest that while the dollar amounts vary normally, the decimal digits do not. The first decimal digit concentrates near 0.8 and 0.9, and the second digit is dominated by 0.09. This pattern indicates that most prices end in .89 or .99, rather than being spread evenly across decimal values. This reflects a common retail pricing strategy where prices are intentionally set just below the next dollar amount because shoppers tend to perceive these prices as meaningfully lower.

7. Run a linear regression of sales on price_int, price_decimal1, price_decimal2, and feat. Report the coefficients of three price variables and explain how to interpret the coefficients.

```python
reg = smf.ols("sales ~ price_int + price_decimal1 + price_decimal2 + feat", data=df).fit()
print(reg.summary())
```

Interpretation of the coefficients:price_int: −13,620

A one-dollar increase in the whole-dollar part of the price is associated with a large drop in sales, averaging around 13,620 units.

price_decimal1: −11,780

Increasing the first decimal digit by 0.1 (for example from .8 to .9) also reduces sales, although by a smaller amount than changing the whole dollar.

price_decimal2: +41,490

The coefficient +41,490 is positive, which is somewhat counterintuitive since raising the price by a cent should normally reduce sales. An explintion for this could be that the model is picking up retail pricing patterns rather than a true causal effect of adding one cent.

Overall, the results show that higher prices reduce sales when looking at the whole-dollar and first-decimal parts of the price, which is consistent with normal demand behavior. The unusual positive effect on the second decimal digit likely reflects pricing conventions in the dataset rather than a real response to very small price changes.
(All statistically significant with p < 0.001.)

9. Split the data into two parts. Run a linear regression of log(sales) on log(price) using part 1. Report the in-sample R2 for part 1 data and calculate the regression model’s out-of-sample R2 for part 2 data.

```python
from sklearn.model_selection import train_test_split
import numpy as np
import statsmodels.formula.api as smf

# split the data
df1, df2 = train_test_split(df, test_size=0.5, random_state=52610)
reg = smf.ols("np.log(sales) ~ np.log(price)", data=df1).fit()
in_samle_r2 = reg.rsquared
in_samle_r2
```

```python
# Predict log(sales) for part 2
y_pred = reg.predict(df2)

# Compute true log(sales) for part 2
y_true = np.log(df2["sales"])

# Compute out-of-sample R2
sse = ((y_true - y_pred)**2).sum()
sst = ((y_true - y_true.mean())**2).sum()
out_of_sample_r2 = 1 - sse/sst
out_of_sample_r2
```

In-sample R2 for part 1: 0.415080940351838
Out-of-sample R2 for part 2: 0.43822604660735254

10. Compare the in-sample and out-of-sample R2, and discuss the result.

The in-sample R² is 0.415080940351838 and the out-of-sample R² is 0.43822604660735254. The two values are very close, which indicates that the simple log–log model fits both halves of the data in a similar way. The slightly higher out-of-sample R² is reasonable in a large dataset where the training and testing subsets come from the same underlying distribution. Overall, the model shows stable performance and does not appear to overfit.
